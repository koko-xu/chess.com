```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
df = pd.DataFrame(columns=["match number", "color", "opponent ranking", "my ranking", "ranking", "level", "moves", "outcome", "termination"])

def to_int(line):
    words = line.split()
    for word in words:
        if '"' in word:
            w = word.translate({ord(i): None for i in '\"\"]'})
            num = int(w)
    return num

file = open('Matches')
match_number = 0

for line in file: 
    # Match Number
    if "[Event \"Live Chess\"]" in line:
        color = ""
        opponent_ranking = 0
        my_ranking = 0
        ranking = ""
        level = ""
        moves = 0
        outcome = ""
        termination = ""
        match_number += 1
        
    # Color
    if "[White \"kokoxu\"]" in line: 
        color = "White"
    elif "[Black \"kokoxu\"]" in line: 
        color = "Black"
    
    # My Ranking, Opponent Ranking
    if "[WhiteElo \"" in line and color == "White":
        num = to_int(line)
        my_ranking = num
    elif "[WhiteElo \"" in line and color == "Black":
        num = to_int(line)
        opponent_ranking = num
    if "[BlackElo \"" in line and color == "Black":
        num = to_int(line)
        my_ranking = num
    elif "[BlackElo \"" in line and color == "White":
        num = to_int(line)
        opponent_ranking = num
    
    # Ranking
    if opponent_ranking > my_ranking:
        ranking = "Over"
    elif opponent_ranking < my_ranking:
        ranking = "Under"
    else: 
        ranking = "Even"
        
    # Level
    if my_ranking in range(600,800):
        level = "G"
    elif my_ranking in range(800,1000):
        level = "F"
    elif my_ranking in range(1000,1200):
        level = "E"
    elif my_ranking in range(1200,1400):
        level = "D"
    elif my_ranking in range(1400,1600):
        level = "C"
    elif my_ranking in range(1600,1800):
        level = "B"
    else:
        level = "-"
    
    # Moves
    moves = 0
    
    # Outcome, Termination
    if "[Termination \"" in line:
        if "kokoxu" in line and "won" in line:
            outcome = "Win"
        elif "drawn" in line:
            outcome = "Draw"
        elif "kokoxu" not in line and "drawn" not in line:
            outcome = "Lose"
        
        words = [s for s in line.split()]
        termination = words[-1]
        
        row = {"match number" : match_number, 
               "color" : color, 
               "opponent ranking" : opponent_ranking, 
               "my ranking" : my_ranking, 
               "ranking" : ranking, 
               "level" : level, 
               "moves" : moves, 
               "outcome" : outcome, 
               "termination" : termination}
        df = df.append(row, ignore_index = True)
        print("Row", match_number, "added")
```

```python
# Confusion Matrix function
def plot_confusion_matrix(cm, classes=None, title='Confusion matrix'):
    """Plots a confusion matrix"""
    if classes is not None:
        sns.heatmap(cm, xticklabels=classes, yticklabels=classes, vmin=0, vmax=1, annot=True, annot_kws={'size':50})
    else: 
        sns.heatmap(cm, vmin=0., vmax=1.)
    plt.title(title)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
y_pred = dt.predict(X_train)

# Plotting Confusion Matrix
cm = confusion_matrix(y_train, y_pred)
cm_norm = cm/cm.sum(axis=1)[:, np.newaxis]
plt.figure()
plot_confusion_matrix(cm_norm, classes=dt.classes_, title='Training confusion')
```

```python
# Calculating False Positives, False Negatives, True Positives, True Negatives
FP = cm.sum(axis=0) - np.diag(cm)
FN = cm.sum(axis=1) - np.diag(cm)
TP = np.diag(cm)
TN = cm.sum() - (FP + FN + TP)

# Sensitivity, hit rate, recall, or true positive rate
TPR = TP/(TP+FN)
print("The True Positive Rate or Recall per class is:", list(TPR))

# Precision or positive predictive value
PPV = TP/(TP+FP)
print("The Precision per class is:", list(PPV))

# False positive rate or False alarm rate
FPR = FP/(FP+TN)
print("The False Alarm rate per class is:", list(FPR))

# False negative rate or Miss rate
FNR = FN/(TP+FN)
print("The Miss rate per class is:", list(FNR))

# Classification error
CER = (FP+FN)/(TP+FP+FN+TN)
print("The Classification Error per class is:", list(CER))

# Overall accuracy
ACC = (TP+TN)/(TP+FP+FN+TN)
print("The Accuracy per class is:", list(ACC))
print("")

# Total averages
print ("The average Recall is: ",TPR.sum()/2)
print ("The average Precision is: ",PPV.sum()/2)
print ("The average False Alarm is: ",FPR.sum()/2)
print ("The average Miss Rate rate is: ",FNR.sum()/2)
print ("The average Classification error is", CER.sum()/2)
print ("The average Accuracy is", ACC.sum()/2)
```
