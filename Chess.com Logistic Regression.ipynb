```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
df = pd.DataFrame(columns=["match number", "color", "opponent ranking", "my ranking", "ranking", "level", "moves", "outcome", "termination"])

def to_int(line):
    words = line.split()
    for word in words:
        if '"' in word:
            w = word.translate({ord(i): None for i in '\"\"]'})
            num = int(w)
    return num

file = open('Matches')
match_number = 0

for line in file: 
    # Match Number
    if "[Event \"Live Chess\"]" in line:
        color = ""
        opponent_ranking = 0
        my_ranking = 0
        ranking = ""
        level = ""
        moves = 0
        outcome = ""
        termination = ""
        match_number += 1
        
    # Color
    if "[White \"kokoxu\"]" in line: 
        color = "White"
    elif "[Black \"kokoxu\"]" in line: 
        color = "Black"
    
    # My Ranking, Opponent Ranking
    if "[WhiteElo \"" in line and color == "White":
        num = to_int(line)
        my_ranking = num
    elif "[WhiteElo \"" in line and color == "Black":
        num = to_int(line)
        opponent_ranking = num
    if "[BlackElo \"" in line and color == "Black":
        num = to_int(line)
        my_ranking = num
    elif "[BlackElo \"" in line and color == "White":
        num = to_int(line)
        opponent_ranking = num
    
    # Ranking
    if opponent_ranking > my_ranking:
        ranking = "Over"
    elif opponent_ranking < my_ranking:
        ranking = "Under"
    else: 
        ranking = "Even"
        
    # Level
    if my_ranking in range(600,800):
        level = "G"
    elif my_ranking in range(800,1000):
        level = "F"
    elif my_ranking in range(1000,1200):
        level = "E"
    elif my_ranking in range(1200,1400):
        level = "D"
    elif my_ranking in range(1400,1600):
        level = "C"
    elif my_ranking in range(1600,1800):
        level = "B"
    else:
        level = "-"
    
    # Moves
    moves = 0
    
    # Outcome, Termination
    if "[Termination \"" in line:
        if "kokoxu" in line and "won" in line:
            outcome = "Win"
        elif "drawn" in line:
            outcome = "Draw"
        elif "kokoxu" not in line and "drawn" not in line:
            outcome = "Lose"
        
        words = [s for s in line.split()]
        termination = words[-1]
        
        row = {"match number" : match_number, 
               "color" : color, 
               "opponent ranking" : opponent_ranking, 
               "my ranking" : my_ranking, 
               "ranking" : ranking, 
               "level" : level, 
               "moves" : moves, 
               "outcome" : outcome, 
               "termination" : termination}
        df = df.append(row, ignore_index = True)
        print("Row", match_number, "added")
```

```python
for val in df["termination"].unique():
    v = val.translate({ord(i): None for i in '\"]'})
    df = df.replace([val], v)
```

```python
color_df = pd.get_dummies(df["color"], drop_first=True)
color_df.head()
```

```python
ranking_df = pd.get_dummies(df["ranking"], drop_first=True)
ranking_df.head()
```

```python
outcome_df = pd.get_dummies(df["outcome"], drop_first=True)
outcome_df.head()
```

```python
termination_df = pd.get_dummies(df["termination"], drop_first=True)
termination_df.head()
```

```python
level_df = pd.get_dummies(df["level"], drop_first=True)
level_df.head()
```

```python
df = pd.concat([df, color_df, ranking_df, level_df, outcome_df, termination_df], axis=1)
```

```python
df.drop(["match number", "opponent ranking", "my ranking", "ranking", "level", "outcome", "termination"], axis=1, inplace=True)
```

```python
df.drop("color", axis=1, inplace=True)
df.head()
```

```python
df.drop(["Lose"], axis=1, inplace=True)
df.head()
```

```python
df.drop(["agreement","checkmate","material","repetition","resignation","rule","stalemate","time"], axis=1, inplace=True)
df.head()
```

```python
X = df.drop("Win", axis=1)
y = df["Win"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, predictions)
```
